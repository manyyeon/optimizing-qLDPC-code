```python
import h5py
import numpy as np
from optimization.experiments_settings import from_edgelist, tanner_graph_to_parity_check_matrix
from collections import Counter

filepath = "optimization/results/fast_greedy_start_from_d_best_2_single_parent_excluding_parents.hdf5"
code_name = "[625,25]"
initial_code_file = "optimization/initial_codes/HGP_(3,4)_[625,25].txt"

def load_initial_code(filepath):
    # Reads the edge list format of the initial code
    with open(filepath, 'r') as f:
        lines = f.readlines()
    num_cn, num_vn = map(int, lines[0].strip().split())
    
    # We will reconstruct H directly
    
    H = np.zeros((num_cn, num_vn), dtype=np.uint8)
    for cn, line in enumerate(lines[1:]):
        neighbors = list(map(int, line.strip().split()))
        for vn in neighbors:
            H[cn, vn] = 1
    return H

def get_variable_regularity(H):
    # Sum column-wise to get the degree (weight) of each variable node
    col_weights = np.sum(H, axis=0)
    # Count how many variable nodes have each degree
    counts = dict(Counter(col_weights))
    return dict(sorted(counts.items()))

def get_check_regularity(H):
    # Sum row-wise to get the degree (weight) of each check node
    row_weights = np.sum(H, axis=1)
    # Count how many check nodes have each degree
    counts = dict(Counter(row_weights))
    return dict(sorted(counts.items()))

# Load Codes
# 1. Initial code
H_initial = load_initial_code(initial_code_file)
v_reg_initial = get_variable_regularity(H_initial)
c_reg_initial = get_check_regularity(H_initial)

# 2. Results from Fast Greedy
with h5py.File(filepath, "r") as f:
    grp = f[code_name]
    lers = grp['logical_error_rates'][:]
    dists = grp['distances_classical'][:]
    states = grp['states'][:]
    
    # Let's find best state according to LER
    best_idx = np.argmin(lers)
    best_ler = lers[best_idx]
    best_dist = dists[best_idx]
    
    best_state = from_edgelist(states[best_idx])
    H_best = tanner_graph_to_parity_check_matrix(best_state)
    H_best = np.asarray(H_best, dtype=np.uint8)
    v_reg_best = get_variable_regularity(H_best)
    c_reg_best = get_check_regularity(H_best)

    # Let's find another distance 8 code with highest LER (worst among d=8)
    higher_ler_indices = np.where(dists == 8)[0]
    other_idx = higher_ler_indices[np.argmax(lers[higher_ler_indices])]
    other_ler = lers[other_idx]
    other_dist = dists[other_idx]
    
    other_state = from_edgelist(states[other_idx])
    H_other = tanner_graph_to_parity_check_matrix(other_state)
    H_other = np.asarray(H_other, dtype=np.uint8)
    v_reg_other = get_variable_regularity(H_other)
    c_reg_other = get_check_regularity(H_other)

print(f"--- Initial Code (d=8) ---")
print(f"Variable Regularity: {v_reg_initial}")
print(f"Check Regularity: {c_reg_initial}")
print(f"H Shape: {H_initial.shape}")

print(f"\n--- Other Code (d={other_dist}, LER={other_ler:.6f}) ---")
print(f"Variable Regularity: {v_reg_other}")
print(f"Check Regularity: {c_reg_other}")
print(f"H Shape: {H_other.shape}")

print(f"\n--- Best Code (d={best_dist}, LER={best_ler:.6f}) ---")
print(f"Variable Regularity: {v_reg_best}")
print(f"Check Regularity: {c_reg_best}")
print(f"H Shape: {H_best.shape}")
```

```python
import h5py
from optimization.analyze_codes.decoder_performance_from_state import evaluate_performance_of_state
from optimization.experiments_settings import from_edgelist
import numpy as np

def load_and_evaluate_code(filepath, code_name=None, run_name=None, state_key="best_state", p=0.03, MC_budget=int(1e6), state_idx=None):
    """
    Loads a code state from an HDF5 result file or a plain text initial code file,
    and evaluates its Logical Error Rate (LER).
    
    If filepath ends with .txt, it assumes it's an initial code text file.
    If it ends with .hdf5, it will fetch the state.
    """
    
    if filepath.endswith('.txt'):
        # For initial code, use the load_initial_code function we made earlier
        H, n_v, n_c = load_initial_code(filepath)
        # Convert H to a MultiGraph state that evaluate_performance_of_state expects
        import networkx as nx
        state = nx.MultiGraph()
        state.add_nodes_from(range(n_c), element_type=0)
        state.add_nodes_from(range(n_c, n_c + n_v), element_type=1)
        for c in range(n_c):
            for v in range(n_v):
                if H[c, v] == 1:
                    state.add_edge(c, n_c + v)
                    
        print(f"Loaded Initial Code from {filepath}")
        code_cost = "Not explicitly recorded"
        
    else:
        # Load from HDF5
        with h5py.File(filepath, "r") as f:
            if code_name is None:
                raise ValueError("code_name must be provided for hdf5 files.")
            grp = f[code_name]
            
            if run_name is not None:
                # If there is a nested run group
                grp = grp[run_name]
                
            if state_key == "states" and state_idx is not None:
                state_edge_list = grp["states"][:][state_idx]
                code_cost = grp["logical_error_rates"][:][state_idx]
            else:
                state_edge_list = grp[state_key][:]
                code_cost = grp.attrs.get("min_cost", "Not found")

        state = from_edgelist(state_edge_list)
        print(f"Loaded {state_key} from {filepath} (code: {code_name}, run: {run_name})")
        print(f"Saved LER: {code_cost}")

    print(f"Evaluating LER at p={p} with MC={MC_budget}...")
    cost_result = evaluate_performance_of_state(
        state=state, p_vals=[p], MC_budget=MC_budget, canskip=False
    )
    
    return state, cost_result

# Example Usage:
# 1. Run for a specific run in a nested HDF5
# state_run, cost_run = load_and_evaluate_code(
#     filepath="optimization/results/start_from_logical_guided.hdf5", 
#     code_name="[1600,64]", 
#     run_name="beam_from_logical_bw2_d10_p0.03_100000screen_1000000prec_exp50_20260412_132444",
#     p=0.03, MC_budget=int(1e5)
# )

# 2. Run for the initial code
# state_init, cost_init = load_and_evaluate_code(
#     filepath="optimization/initial_codes/HGP_(3,4)_[1600,64].txt",
#     p=0.03, MC_budget=int(1e5)
# )
```
