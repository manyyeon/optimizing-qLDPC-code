```python
import h5py

src_path = "optimization/results/logical_guided_search_test.hdf5"
dst_path = "optimization/results/logical_guided_search.hdf5"


def copy_group(src_group, dst_group):
    # Copy attributes
    for key, val in src_group.attrs.items():
        if key not in dst_group.attrs:
            dst_group.attrs[key] = val

    for name, item in src_group.items():
        if isinstance(item, h5py.Dataset):
            if name not in dst_group:
                src_group.copy(name, dst_group)
            else:
                print(f"Skip dataset (exists): {dst_group.name}/{name}")

        elif isinstance(item, h5py.Group):
            if name not in dst_group:
                new_grp = dst_group.create_group(name)
            else:
                new_grp = dst_group[name]

            copy_group(item, new_grp)


with h5py.File(src_path, "r") as src, h5py.File(dst_path, "a") as dst:
    for code_name in src:
        if code_name not in dst:
            print(f"Creating code group: {code_name}")
            dst.create_group(code_name)

        src_code_grp = src[code_name]
        dst_code_grp = dst[code_name]

        for run_name in src_code_grp:
            if run_name not in dst_code_grp:
                print(f"Copying run: {code_name}/{run_name}")
                src_code_grp.copy(run_name, dst_code_grp)
            else:
                print(f"Run exists, merging inside: {code_name}/{run_name}")
                copy_group(src_code_grp[run_name], dst_code_grp[run_name])
```
